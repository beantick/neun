Buatlah program yang bisa mengklasifikasikan angka genap dan ganjil dari inputan user.

Contoh:

- Input: 
        5

- Output:
        5 adalah angka ganjil

**3.**
market if.PNG

```python
# Insert your answer here

jumlah_apel = int(input('Masukkan jumlah apel: '))
jumlah_jeruk = int(input('Masukkan jumlah jeruk: '))
jumlah_anggur = int(input('Masukkan jumlah anggur: '))

# Diketahui harga buah sebagai berikut:
harga_apel = 10000
harga_jeruk = 15000
harga_anggur = 20000

print('\nDetail Belanja')

print(f'Apel: \t {jumlah_apel} x {harga_apel} = {jumlah_apel*harga_apel}')
print(f'Jeruk: \t {jumlah_jeruk} x {harga_jeruk} = {jumlah_jeruk*harga_jeruk}')
print(f'Anggur:\t {jumlah_anggur} x {harga_anggur} = {jumlah_anggur*harga_anggur}')

total_belanja = (jumlah_apel*harga_apel) + (jumlah_jeruk*harga_jeruk) + (jumlah_anggur*harga_anggur)
print(f'\nTotal belanjaan Anda Rp {total_belanja}')

bayar = int(input('\nMasukkan jumlah uang: '))

if bayar > total_belanja:
  print(f'\nUang kembalian Anda: Rp {bayar - total_belanja}')
  print('\nTerima kasih')
elif bayar == total_belanja:
  print('\nTerima kasih')
else:
  print('\nTransaksi Anda dibatalkan')
  print(f'\nUang Anda kurang sebesar Rp {total_belanja - bayar}')
```

Jika x adalah bilangan puluhan, munculkan 'x adalah puluhan'.

```python
# Insert your answer here

x = int(input('Masukkan angka antara 1 hingga 99: '))

if x in range(10, 100):
  print(f'{x} adalah puluhan')

  if x%2 == 0:
    print(f'{x} adalah bilangan genap')
  else:
    print(f'{x} adalah bilangan ganjil')

elif x < 10 and x >= 1:
  print(f'{x} adalah satuan')

else:
  print('Angka invalid. Silakan masukkan angka yang sesuai.')
```

Buatlah sebuah variable yang menampung sebuah list berisi macam-macam nilai yang memiliki tipe data berbeda.

- Contoh:
      values = [True, 13, 'Hello', 2.7]
 
- Buatlah program untuk mengetahui tipe data dari tiap item.

- Gunakan user input yang nantinya akan diisi index dari values. 

Contoh: ('Masukkan index:')

- Contoh output:
      True bertipe data boolean
      2.7 bertipe data float

```python
# Insert your answer here

values = [True, 13, 'Hello', 2.7]
user_input = int(input('Masukkan angka index yang ingin diketahui tipe datanya: '))

if user_input > len(values) - 1:
  print('Angka index tidak ditemukan.')
elif type(values[user_input]) == str:
  print(f'{values[user_input]} bertipe data string')
elif type(values[user_input]) == int:
  print(f'{values[user_input]} bertipe data integer')
elif type(values[user_input]) == float:
  print(f'{values[user_input]} bertipe data float')
elif type(values[user_input]) == bool:
  print(f'{values[user_input]} bertipe data boolean')
  
```

**10.**

Buatlah program yang dapat menentukan angka manakah yang nominalnya paling besar dari 3 angka yang diinput oleh user.

User bebas memasukkan angka berapapun.

Contoh:

- Input:
      10
      23
      42


- Output:
      42 adalah angka terbesar dari semua angka yang diinput.

```python
# Insert your answer here

angka_pertama = int(input('Masukkan angka pertama: '))
angka_kedua = int(input('Masukkan angka kedua: '))
angka_ketiga = int(input('Masukkan angka ketiga: '))

if angka_pertama > angka_kedua and angka_pertama > angka_ketiga:
  print(f'{angka_pertama} adalah angka terbesar dari semua angka yang diinput.')
elif angka_kedua > angka_pertama and angka_kedua > angka_ketiga:
  print(f'{angka_kedua} adalah angka terbesar dari semua angka yang diinput.')
elif angka_ketiga > angka_pertama and angka_ketiga > angka_kedua:
  print(f'{angka_ketiga} adalah angka terbesar dari semua angka yang diinput.')
else:
  print('Ketiga angka bernilai sama')
```

**11.**

Dari list berikut: [12, 15, 1, 7, 4, 100]

Buatlah program untuk mencari angka tertinggi di dalam list tanpa menggunakan fungsi max atau sort pada Python.

- Output:
      100 

```python
my_list = [12, 15, 1, 7, 4, 100]

angka_tertinggi = my_list[0]

for angka in my_list:
    if angka > angka_tertinggi:
        angka_tertinggi = angka

print(angka_tertinggi)
```
